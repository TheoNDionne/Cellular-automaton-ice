```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

params = {
    "xtick.top": True, 
    "ytick.right": True, 
    "xtick.direction": "in", 
    "ytick.direction": "in", 
    "figure.dpi": 200
    }
plt.rcParams.update(params)

sns.set_palette("plasma")
```

```python
def attachment_coefficient(sigma, k, A=1, b=1):
    return A*k*np.exp(-b/(k*sigma))

sat_space = np.linspace(0,1,300)
```

```python
plt.figure()

max_attachment_coef = attachment_coefficient(1,5,b=10)

for k in range(1,6):
    plt.plot(sat_space, attachment_coefficient(sat_space, k, b=10)/max_attachment_coef, label=f"k={k}")

plt.legend()

plt.xlabel("$\sigma$")
plt.ylabel("$\\alpha(\sigma)\,/\,\\alpha_\operatorname{max}$")

plt.show()
```

```python
plt.figure()

max_attachment_coef = attachment_coefficient(1,5,b=1)

for k in range(1,6):
    plt.plot(sat_space, attachment_coefficient(sat_space, k, b=1)/max_attachment_coef, label=f"k={k}")

plt.legend()

plt.xlabel("$\sigma$")
plt.ylabel("$\\alpha(\sigma)\,/\,\\alpha_\operatorname{max}$")

plt.show()
```

```python
plt.figure()

max_attachment_coef = attachment_coefficient(1,5,b=0.1)

for k in range(1,6):
    plt.plot(sat_space, attachment_coefficient(sat_space, k, b=0.1)/max_attachment_coef, label=f"k={k}")

plt.legend()

plt.xlabel("$\sigma$")
plt.ylabel("$\\alpha(\sigma)\,/\,\\alpha_\operatorname{max}$")

plt.show()
```
